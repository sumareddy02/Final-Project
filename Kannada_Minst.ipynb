```python

import numpy as np, pandas as pd

#Loading train and test dataset
x_train = np.load('E:/DataScience/X_kannada_MNIST_train.npz')['arr_0']
x_test = np.load('E:/DataScience/X_kannada_MNIST_test.npz')['arr_0']
y_train = np.load('E:/DataScience/y_kannada_MNIST_train.npz')['arr_0']
y_test = np.load('E:/DataScience/y_kannada_MNIST_test.npz')['arr_0']

print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.decomposition import PCA

x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_test_flat = x_test.reshape(x_test.shape[0], -1)

n_components = 10
pca = PCA(n_components=n_components)

x_train_pca = pca.fit_transform(x_train_flat)
x_test_pca = pca.transform(x_test_flat)

print(x_test_pca.shape)
print(x_train_pca.shape)
```

```python
#KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix

model=KNeighborsClassifier(n_neighbors=2)
model.fit(x_train_pca,y_train)
train_pred = model.predict(x_train_pca)
test_pred = model.predict(x_test_pca)

print(f"*****{type(model).__name__}")
print(f"Train Precision: {precision_score(y_train,train_pred,average='micro')}")
print(f"Train Recall: {recall_score(y_train,train_pred,average='micro')}")
print(f"Train F1 Score: {f1_score(y_train,train_pred,average='micro')}")
print("Confusion Matrix:")
print(confusion_matrix(y_test,test_pred))
print("Test Data")
print(f"Test Precision: {precision_score(y_test,test_pred,average='micro')}")
print(f"Test Recall: {recall_score(y_test,test_pred,average='micro')}")
print(f"Test F1 Score: {f1_score(y_test,test_pred,average='micro')}")
print("Confusion Matrix:")
print(confusion_matrix(y_test,test_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix

model=DecisionTreeClassifier(criterion="entropy",max_depth=3,min_samples_split=10)
model.fit(x_train_pca,y_train)
train_pred = model.predict(x_train_pca)
test_pred = model.predict(x_test_pca)

print(f"*****{type(model).__name__}")
print(f"Train Precision: {precision_score(y_train,train_pred,average='micro')}")
print(f"Train Recall: {recall_score(y_train,train_pred,average='micro')}")
print(f"Train F1 Score: {f1_score(y_train,train_pred,average='micro')}")
print("Confusion Matrix:")
print(confusion_matrix(y_test,test_pred))
print("Test Data")
print(f"Test Precision: {precision_score(y_test,test_pred,average='micro')}")
print(f"Test Recall: {recall_score(y_test,test_pred,average='micro')}")
print(f"Test F1 Score: {f1_score(y_test,test_pred,average='micro')}")
print("Confusion Matrix:")
print(confusion_matrix(y_test,test_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix

model=RandomForestClassifier()
model.fit(x_train_pca,y_train)
train_pred = model.predict(x_train_pca)
test_pred = model.predict(x_test_pca)

print(f"*****{type(model).__name__}")
print(f"Train Precision: {precision_score(y_train,train_pred,average='micro')}")
print(f"Train Recall: {recall_score(y_train,train_pred,average='micro')}")
print(f"Train F1 Score: {f1_score(y_train,train_pred,average='micro')}")
print("Confusion Matrix:")
print(confusion_matrix(y_test,test_pred))
print("Test Data")
print(f"Test Precision: {precision_score(y_test,test_pred,average='micro')}")
print(f"Test Recall: {recall_score(y_test,test_pred,average='micro')}")
print(f"Test F1 Score: {f1_score(y_test,test_pred,average='micro')}")
print("Confusion Matrix:")
print(confusion_matrix(y_test,test_pred))
```
